BB84 프로토콜의 목표는 Alice와 Bob 사이에 비밀 키를 만들어서 두 사람이 숨겨진 메시지를 암호화하고 해독하는 데 사용하는 것입니다. 이 연습에서는 암호화된 메시지를 해독하기 위한 비밀 키를 생성하기 위한 프로토콜의 여러 단계를 안내합니다.


## BB84 프로토콜

BB84 프로토콜의 단계를 살펴 보겠습니다.

1. 첫 번째 단계에서 Alice는 $ n $ 비트로 구성되는 두 개의 임의 비트 문자열 $ k $ 및 $ b $를 선택합니다. 그녀의 비트 문자열 $ k $에는 인코딩하려는 실제 비트가 포함되며 (그 암호키는 나중에 작성됩니다) $ b $는 비트를 인코딩 할 기준을 결정합니다. $ b_i = 0 $의 경우 (즉, $ i $번째 비트가 0 인 경우) $ \{| 0 \rangle, | 1 \rangle \}$를 베이스로 $i$번째 큐빗를 인코딩합니다. 반면, $ b_i = 1 $의 경우 $ \{| + \rangle, |-\rangle \}$세트를 베이스로 $i$번째 큐빗을 인코딩하는데, 이때 $| + \rangle : = \frac {1} {\sqrt {2}} (| 0 \rangle + | 1 \rangle) \}$, $ |-\rangle : = \frac {1} {\sqrt {2}} (| 0 \rangle-| 1 \rangle) $ 입니다. 
이 서로 다른 두 베이스는 $ 45 ^ \circ $만큼 회전되는 두 개의 수직 화살표로 이해할 수 있습니다.
따라서 각 큐빗 $ q_i $의 인코딩은 다음과 같습니다.

<img src = "encoding_Alice.png" alt = "drawing" width = "300"/>

2. $ n $개의 큐빗을 인코딩 한 후 Alice는 이 큐빗들을 Bob에게 보냅니다. 또한 Bob은 측정할 베이스를 결정하는 $ n $ 비트로 구성된 임의 비트 문자열 $ \tilde {b} $를 선택합니다. 그는 측정 결과 $ \tilde {k_i} $를 해당 기준 비트 $ \tilde {b_i} $와 함께 테이블에 저장합니다.

3. 다음으로 Alice와 Bob은 각자의 기준 비트 $ b_i $와 $ \tilde{b}_i $를 비교합니다. Bob은 $ b_i \neq \tilde{b}_i $ 인 $i$에 대하여 Alice의 큐빗이 인코딩 된 것과 다른 방식으로 측정하였기 때문에 $ \frac {1} {2} $ 확률로 각 결과를 얻었을 것입니다(역주:결과가 완전히 무작위입니다). 따라서 Alice와 Bob은 이러한 기준 비트에 해당하는 모든 키 비트를 폐기합니다. 그러나 $ b_i = \tilde {b}_i $ 인 경우 동일한 기준으로 큐 비트를 준비하고 측정 했으므로 (누군가 도청하지 않는 한) Bob은 Alice가 인코딩한 키 비트를 얻을 것입니다. 이렇게 얻은 결과로 암호키를 만들어 냅니다.

이 예제에서 생성 된 키는 '0110'입니다. 키가 비밀스럽고 정확한지 확인하기 위해 Alice와 Bob은 키 비트 중 일부를 "희생"하여 도청하지 않았는지 확인합니다. 누군가가 도중에 큐빗를 측정했다면, 그 큐빗의 상태가 변경되었을 가능성이 있으며 $ \frac{1}{4} $로 Bob과 Alice의 키 비트가 달라집니다. $ m $ 비트를 확인하면 도청자가 눈에 띄지 않을 확률이 $ \left(\frac{3}{4} \right)^m $로 줄어 듭니다. 따라서 충분한 비트를 확인하고 모두 동일하면 도청자가 없고 키가 비밀이라고 가정 할 수 있습니다. 그러나 이번 연습문제에서는 문제를 단순하게 하기 위해 이러한 테스트를 수행하지 않습니다. 대신, 키의 모든 비트가 사용됩니다.

### 메시지 암호화
비밀 키가 배포되면 Alice는 소위 [one-time pad] (https://en.wikipedia.org/wiki/One-time_pad) 기술을 사용하여 메시지를 암호화 할 수 있습니다: 그녀가 보내고 싶은 비밀 메시지 비트에 키 비트를 더하는 방법입니다. 위의 예에서 키는 $ \text{key} =`0110` $입니다. 비밀 메시지 비트 문자열이 $ m =`1100` $이면 암호화 된 메시지는 $ c = m \oplus \text{key} \mod 2 =`1010` $입니다. 그런 다음 Bob은 암호화 된 메시지에 키를 추가하여 ($ m = c \oplus \text {key} \mod 2 $) 메시지를 해독 할 수 있습니다.
